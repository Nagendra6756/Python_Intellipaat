```python
data1=pd.read_csv('mtcars.csv')
data1
```

```python
data1=data1.fillna({
    'Unnamed: 1':'No name',
    'disp':data1['disp'].mean(),
    'hp':data1['hp'].median()
})
```

```python
data1
```

```python
data1=data1.fillna(method='bfill')
```

```python
data1
```

```python
data1=data1.dropna(how='all') ---removes rows when entire row is nan
#data1=data1.dropna(threshold=5)  --removes row when data in a row is <4
```

```python
data1
```

```python
data1['carb']=data1['carb']*2
data1
```

```python
def double(num):
    return num * 2

data1['card']=data1['carb'].apply(double)
```

```python
data1
```

```python
def capitalize(string):
    return string.upper()
data1['Unnamed: 1']=data1['Unnamed: 1'].apply(capitalize)
data1
```

```python
#filtering
cylinder_greater_5=data1['cyl']>5
cylinder_greater_5
```

```python
#joining or merging 
a=pd.DataFrame({
    'color':['red','green','yellow'],
    'stock':[10,22,33],
    'places':['chennai','delhi','punjab']
})
b=pd.DataFrame({
    'color':['yellow','green','pink'],
    'sizes':['small','medium','large'],
    'places':['delhi','punjab','chennai']
})
```

```python
pd.merge(a,b,on='color')
```

```python
pd.merge(a,b,on='places')
```

```python
a=pd.DataFrame({
    'color':['red','green','yellow'],
    'stock':[10,22,33]
})
b=pd.DataFrame({
    'color':['yellow','green','pink'],
    'sizes':['small','medium','large']
})
```

```python
pd.merge(a,b,how='inner')
```

```python
pd.merge(a,b,how='outer')
```

```python
#Matplotlib
#figsize should be set first
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,5),dpi=500) 
x=[3,4,5,5.5,7]
y=[30,20,50,40,60]
plt.plot(x,y,linestyle='-',linewidth=1,marker='.')
plt.title('Data points', fontdict={'fontname':'Times New Roman','fontsize':20})
plt.xlabel('experience')
plt.ylabel('salary in lakhs')
plt.xticks([3,4,5,6,7])
plt.yticks(range(20,100,10))

```
