```python
def add(n1,n2):
    return n1+n2
def sub(n1,n2):
    return n1-n2
def div(n1,n2):
    return n1/n2
def operator(n1,n2,operation):
    value=operation(n1,n2)
    return value
print(operator(5,3,add))
print(operator(10,3,sub))
```

```python
nums=[1,2,3,4,5,6,7,8]
def get_factorial(num):
    fact=1
    for i in range(2,num+1):
        fact=fact*i
    return fact

fact_nums=[]
for num in nums:
    factorial=get_factorial(num)
    fact_nums.append(factorial)
print(fact_nums)
```

```python
nums=[1,2,3,4,5,6,7,8]
def even_num(num):
    return num%2==0
even_num=filter(even_num,nums)
print(list(even_num))
```

```python
fig,axes=plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.scatterplot(x=df['hp'],y=df['mpg'])
plt.title('Hp with mpg')
plt.ylabel('Miles per gallon')
plt.xlabel('Horsepower')

plt.subplot(2,2,2)
sns.scatterplot(x=df['hp'],y=df['time-to-60'])
plt.title('Hp with time-to-60')
plt.ylabel('time-to-60')
plt.xlabel('Horsepower')

plt.subplot(2,2,3)
sns.scatterplot(x=df['hp'],y=df['weightlbs'])
plt.title('Hp with weightlbs')
plt.ylabel('weightlbs')
plt.xlabel('Horsepower')

plt.subplot(2,2,4)
sns.scatterplot(x=df['hp'],y=df['cylinders'])
plt.title('Hp with cylinders')
plt.ylabel('cylinders')
plt.xlabel('Horsepower')
```

```python
fig,axes=plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.scatterplot(x=df['hp'],y=df['mpg'],hue=df['brand'])
plt.title('Hp with mpg')
plt.ylabel('Miles per gallon')
plt.xlabel('Horsepower')

plt.subplot(2,2,2)
sns.scatterplot(x=df['hp'],y=df['time-to-60'],hue=df['brand'])
plt.title('Hp with time-to-60')
plt.ylabel('time-to-60')
plt.xlabel('Horsepower')

plt.subplot(2,2,3)
sns.scatterplot(x=df['hp'],y=df['weightlbs'],hue=df['brand'])
plt.title('Hp with weightlbs')
plt.ylabel('weightlbs')
plt.xlabel('Horsepower')

plt.subplot(2,2,4)
sns.scatterplot(x=df['hp'],y=df['cylinders'],hue=df['brand'])
plt.title('Hp with cylinders')
plt.ylabel('cylinders')
plt.xlabel('Horsepower')
```
